# Line plot

Two random walks

```python
import random

walk_1 = [0]
walk_2 = [0]
steps = 1000

for i in range(steps):
    change = 1 if random.randint(0, 1) else -1
    walk_1.append(walk_1[i] + change)
    walk_2.append(walk_2[i] + np.random.normal())
    
plt.plot(walk_1)
plt.plot(walk_2)
plt.show()
```

# Scatter plot

This uses a color map based on value in the clusters. These make use of the 'sklearn' library, and may need to be installed if you haven't already installed it. 

```python
from sklearn.cluster import KMeans
from sklearn import datasets

blobs_data = []
cluster_data = []
for _ in range(4):
    blobs = datasets.make_blobs(1000)[0]
    blobs_data.append(blobs)
    
    clusters = KMeans(n_clusters=3).fit_predict(blobs)
    cluster_data.append(clusters)

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
for idx, (blobs, clusters) in enumerate(zip(blobs_data, cluster_data)):
    x, y = idx // 2, idx % 2
    X, Y = blobs[:, 0], blobs[:, 1]
    ax[x, y].scatter(X, Y, c=clusters, cmap=ow_colormap());

plt.show()
```

**We have created various colormaps using OW colors**
