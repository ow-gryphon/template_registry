```python
# importing important libs
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set up matplotlib style
mpl_style = Path.cwd().parent / "resources" / "ow_matplotlib_style.mplstyle"

plt.style.use(mpl_style)
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

blobs_data = []
cluster_data = []
for _ in range(4):
    blobs = make_blobs(1000)[0]
    blobs_data.append(blobs)
    
    clusters = KMeans(n_clusters=3).fit_predict(blobs)
    cluster_data.append(clusters)
    

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
for idx, (blobs, clusters) in enumerate(zip(blobs_data, cluster_data)):
    x, y = idx // 2, idx % 2
    X, Y = blobs[:, 0], blobs[:, 1]
    ax[x, y].scatter(X, Y, c=clusters);

plt.show()
```
