```python
# Obtain sample data
size = 50
white_noise = np.random.randn(size)
plt.plot(white_noise)
random_walk = np.cumsum(np.random.randn(size))
plt.plot(random_walk)

temp_rand = np.random.randn(size)
rho = 0.8
ar_series = np.array([0.0 for i in range(0,size)])
ar_series[0] = 0
for i in range(1,size):
    ar_series[i] = ar_series[i-1]*rho + np.sqrt(1-np.power(rho,2))*temp_rand[i]
plt.plot(ar_series)
```

```python
weights = np.power(white_noise, 2)

# Place data into a pandas DataFrame
used_data = pd.DataFrame(data=np.transpose(np.array([white_noise+10, random_walk, ar_series, weights])), columns=["noise", "random", "ar", "weights"])
print(used_data.head())

```
