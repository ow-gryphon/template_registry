## Overview

This notebook provides examples of generating diagnostics plots to better understand the performance of a model. We cover plots that are not generated by standard libraries. These include:
* Actual vs. Fitted plot, where data is bucketed based on the predicted value, and plotted as a scatter plot (actual on y-axis, predicted on x-axis). The scatter plot for a strong model will fall near the 45 degree line. The size of the bucketing can be changed to get a sense of the variability of model errors
* Actual and Fitted vs. X-variables (can be independent variable in model or any other variable of interest). This is created by bucketing the data based on the X-variable value (by percentiles for continuous data, by individual value for discrete/categorical x-variable). This is useful for:
    - Examining non-linearities with existing model variables
    - Checing appropriate transformation of variable in model (e.g. winsorized x-variable used in model, for plotting you use the original x-variable)
    - Checking that relationship with intuitive variables not included in model are still captured (due to correlations with variables included in the model)
    - Checking whether a variable should be added to the model
* Errors vs. X-variables by segment. This is useful for analyzing potential interaction terms
