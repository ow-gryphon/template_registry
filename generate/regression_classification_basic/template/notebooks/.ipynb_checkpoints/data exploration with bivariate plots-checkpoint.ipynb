# Bivariate plots for regression model - Boston dataset

Use the **bivariate_continuous** function within the bivariate_plots module to generate the Bivariate Plot for numerical independent variables. Dependent variable needs to be numerical as well.

Most important options are:
* Number of buckets with which to group the data (bucketed based on independent variable value). 
* Whether to show the number of observations in each bucket (useful if you have high concentration in certain IV values)
* Whether to use confidence intervals or standard deviations

```python
help(bivariate_plots.bivariate_continuous)
```

## Numerical & continuous independent variables

```python
# Basic plot
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", with_count = False, header = "Basic plot")

# With observations and trendline
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", with_count = True, trendline = True,
                                            header = "# Obs and Trendline")

# With log(X) axis
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", with_count = True, trendline = True,
                                            header = "Log of X", x_scale='log')

# With log(Y) and log(X) axis
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", with_count = True, trendline = True,
                                            header = "Log of Y & X", x_scale='log', y_scale='log')

# Basic Plot with standard error
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", 
                                            header = "Standard Errors", with_stderr = True
                                            )

# Plot with standard error
_, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                            y_var = "target", x_var = "DIS", with_count = True, trendline = True,
                                            header = "Standard Error Log X & Y", x_scale='log', y_scale='log', with_stderr = True
                                            )
```

```python
# Since these functions return the figure, you can modify the figure as needed

figure, _ = bivariate_plots.bivariate_continuous(used_data = df_boston, 
                                                 y_var = "target", x_var = "DIS", with_count = False, header = "Basic plot")

figure.suptitle("Changing the title")
figure.set_size_inches(18.5, 10.5) # Change size
figure.savefig('test_figure.png', dpi=100)

ax1 = figure.axes[0] # Getting the axis
ax1.set_ylabel("New Y label")
ax1.set_xlabel("New X label")
```

## Categorical or Discrete Independent Variables

**For numberical & continuous independent variables, we group the data based on percentiles of the independent variable value. Meanwhile for Categorical or Discrete Independent Variables, we bucket based on unique values of the variable.**

The distinction between categorical and discrete, is that the discrete variables can be sorted meaningfully

```python
## CATEGORICAL (Not discrete numerical) ##

# Basic plot
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "age_group", discrete=False,
                                             with_count = False, header = "Basic plot")

# With log(Y) axis
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "age_group", discrete=False,
                                            with_count = True, header = "Log of Y", y_scale='log')

# Basic Plot with standard error
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "age_group", discrete=False,
                                            header = "Standard Errors", with_stderr = True
                                            )

# Plot with standard error
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "age_group", discrete=False,
                                            with_count = True, header = "Standard Error Log Y", y_scale='log', with_stderr = True
                                            )
```

```python
# Since these functions return the figure, you can modify the figure as needed

figure, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "age_group", discrete=False,
                                             with_count = False, header = "Basic plot")

figure.suptitle("Changing the title")
figure.set_size_inches(18.5, 10.5) # Change size
figure.savefig('test_figure.png', dpi=100)

ax1 = figure.axes[0] # Getting the axis
ax1.set_ylabel("New Y label")
ax1.set_xlabel("New X label")
```

```python
## DISCRETE NUMERICAL ##

# Basic plot
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "RAD", discrete=True,
                                             with_count = False, header = "Basic plot")

# With log(Y) axis
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "RAD", discrete=True,
                                            with_count = True, header = "Log of Y", y_scale='log')

# Basic Plot with standard error
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "RAD", discrete=True,
                                            header = "Standard Errors", with_stderr = True
                                            )

# Plot with standard error
_, _ = bivariate_plots.bivariate_categorical(used_data = df_boston, 
                                            y_var = "target", x_var = "RAD", discrete=True,
                                            with_count = True, header = "Standard Error Log Y", y_scale='log', with_stderr = True
                                            )
```
