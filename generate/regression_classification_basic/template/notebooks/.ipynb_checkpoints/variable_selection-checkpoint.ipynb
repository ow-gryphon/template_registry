## Stepwise selection using sklearn's feature selection module:

https://scikit-learn.org/stable/modules/feature_selection.html

Basic functionality in sklearn. Recommended that you use mlxtend (see next section)

```python
from sklearn.feature_selection import RFE, SelectFromModel, SequentialFeatureSelector
```

```python
# Set up for OLS

DV='target'
IVs = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
model_dataset = df_boston[[DV] + IVs].dropna()

Y = model_dataset[DV]
X = model_dataset[IVs]

lr = sklearn.linear_model.LinearRegression()
```

```python
# Recursive feature elimination - OLS

selector = RFE(lr, n_features_to_select=5, step=1)
selector = selector.fit(X, Y)

selected_IV = list(compress(IVs, selector.support_))
selected_IV
```

```python
# Sequential feature selection - OLS

sfs = SequentialFeatureSelector(lr, n_features_to_select=5)
sfs.fit(X, Y)

selected_IV = list(compress(IVs, sfs.support_))
selected_IV
```

## Stepwise selection using mlxtend

This uses the mlxtend library, which is pip-installable. Read more here: http://rasbt.github.io/mlxtend/user_guide/feature_selection/SequentialFeatureSelector/

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
```

```python
# OLS Regression - Forward selection

DV='target'
IVs = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
model_dataset = df_boston[[DV] + IVs].dropna()

lr = sklearn.linear_model.LinearRegression()

sfs = SFS(lr, 
          k_features=10, 
          forward=True, 
          floating=False, 
          scoring='neg_mean_squared_error',
          cv=10)

sfs = sfs.fit(model_dataset[IVs], model_dataset[DV])
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

pd.DataFrame.from_dict(sfs.get_metric_dict()).T
```

```python
# OLS Regression - Backward selection

DV='target'
IVs = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
model_dataset = df_boston[[DV] + IVs].dropna()

lr = sklearn.linear_model.LinearRegression()

sfs = SFS(lr, 
          k_features=3, 
          forward=False, 
          floating=False, 
          scoring='neg_mean_squared_error',
          cv=10)

sfs = sfs.fit(model_dataset[IVs], model_dataset[DV])
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

pd.DataFrame.from_dict(sfs.get_metric_dict()).T
```

**NOTE: For this, you can activate floating = True, to give a proper stepwise back and forth approach similar to what is offered in SAS.**

The floating variants, SFFS and SBFS, can be considered as extensions to the simpler SFS and SBS algorithms. The floating algorithms have an additional exclusion or inclusion step to remove features once they were included (or excluded), so that a larger number of feature subset combinations can be sampled. It is important to emphasize that this step is conditional and only occurs if the resulting feature subset is assessed as "better" by the criterion function after removal (or addition) of a particular feature. Furthermore, I added an optional check to skip the conditional exclusion steps if the algorithm gets stuck in cycles.

```python
# Logistic Regression - Forward selection
dataset = df_cancer
DV = 'target'

IVs = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension']

model_dataset = dataset[[DV] + IVs].dropna()

lr = sklearn.linear_model.LogisticRegression(C=1, solver='liblinear')

sfs = SFS(lr, 
          k_features=10, 
          forward=True, 
          floating=False, 
          scoring='roc_auc',
          cv=10)

sfs = sfs.fit(model_dataset[IVs], model_dataset[DV])
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

pd.DataFrame.from_dict(sfs.get_metric_dict()).T
```
